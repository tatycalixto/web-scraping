```python
from bs4 import BeautifulSoup as bs
import requests as rq
import re
import pandas as pd
import numpy as np
```

```python
resposta = rq.get('https://www.glassdoor.com.br/Sal%C3%A1rios/cientista-de-dados-sal%C3%A1rio-SRCH_KO0,18.htm')
#resposta.text
```

```python
texto = resposta.text
texto_formatado = bs(texto,'html.parser')
texto_formatado.prettify()
```

```python
#lista_empresas = texto_formatado.find_all('h3',{'data-test':'salaries-list-item-0-employer-name'})
lista_empresas = texto_formatado.find_all('h3',{'data-test':re.compile ('salaries-list-item-.*-employer-name')}) #com expressão regular
lista_empresas
```

```python
#salario = texto_formatado.find_all('div',{'data-test':'salaries-list-item-0-salary-info'}) 
lista_salario = texto_formatado.find_all('div',{'data-test':re.compile ('salaries-list-item-.*-salary-info')}) 

h3_salario = lista_salario[0].contents[0].text
h3_salario.replace('R$', '').replace('\xa0','')
```

```python
for salario_ in lista_salario:
  str_salario = salario_.contents[0].text
  print(str_salario)
```

```python
for empresa,salario_ in zip (lista_empresas,lista_salario):
  nome_empresa = empresa.find('a').text
  str_salario = salario_.contents[0].text
  print(f'{nome_empresa} - {str_salario}')
```

```python
lista_todos_salarios = []
for empresa,salario_ in zip (lista_empresas,lista_salario):
  nome_empresa = empresa.find('a').text
  str_salario = salario_.contents[0].text
  str_salario = str_salario.replace('R$', '').replace('\xa0','')

  lista_todos_salarios.append((nome_empresa,str_salario))
lista_todos_salarios
```

```python
from os import listxattr
df_salarios = pd.DataFrame(lista_todos_salarios,columns=['empresa','salario'])
df_salarios['salario'] = df_salarios['salario'].astype(np.float32)
df_salarios
```
